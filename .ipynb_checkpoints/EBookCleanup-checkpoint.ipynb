```python
import re
```

```python
#open file
raw_content = open("lifedivine.txt", 'r')
```

```python
def removeTitleAndChapterHeader(content):
    text = ""
    for line in content:
        matchObj = re.search("(^[\s]*[0-9]+\s.*)|(.*\s[0-9]+[\s]*$)|(^[\s]*Chapter [MDCLXVI]+$)|(^CONTENTS)|(^[\s]*PART [MDCLXVI]+$)", line)
        if matchObj is not None:
            pass
        else:
            text += line

    return text

def fixHyphenatedWords(content):
    pattern = re.compile(r"\s\w+-\s\w+")
    items = re.findall(pattern, content)
    for item in items:
        print item
        content = content.replace(item, item.replace('- ',''))
    return content

def writeToFile(content, fileName):
    f = open(fileName,'w')
    f.write(content)
    f.close
```

```python
cleaned_text = removeTitleAndChapterHeader(raw_content)
cleaned_text = fixHyphenatedWords(cleaned_text)

writeToFile(cleaned_text, "lifedivine_cleaned.txt" )
```
